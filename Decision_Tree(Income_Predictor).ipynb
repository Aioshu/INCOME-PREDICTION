# Data Preparation

Decision trees,  can process categorical variables easily. However, we still need to **encode the categorical variables** into a standard format so that sklearn can understand them and build the tree. We'll do that using the ```LabelEncoder()``` class, which comes with ```sklearn.preprocessing```.

````python
from sklearn import preprocessing
````

````python
income_data_categorical=income_data.select_dtypes(include="object")
income_data_categorical.head()

# applying label encoder to income_data_categorical

le=preprocessing.LabelEncoder()
income_data_categorical=income_data_categorical.apply(le.fit_transform)
income_data_categorical.head()
````

````python
# concatinate income_data_categorical and Original income_data

income_data=income_data.drop(income_data_categorical.columns,axis=1)
income_data=pd.concat([income_data,income_data_categorical],axis=1)

income_data.head()
````

````python
income_data.info()
````

````python
income_data.income.unique()
````

````python
# convert target variable income to categorical
income_data["income"]=income_data["income"].astype("category")
````

````python
income_data.info()
````

# Model Building and Evalution

````python
from sklearn.model_selection import train_test_split
````

````python
X=income_data.drop("income",axis=1)
y=income_data.income
````

````python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=99)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
````

````python
# Importing decision tree classifier
from sklearn.tree import DecisionTreeClassifier

# fitting the Decision tree with default hyperparameters , and max depth which is 5.
DT_Default=DecisionTreeClassifier(max_depth=5)
DT_Default.fit(X_train,y_train)
````

````python
# Checking the accuracy 
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix

# predictions
y_pred=DT_Default.predict(X_test)


# Printing classification_report
print(classification_report(y_test,y_pred))
````

````python
# printing  confusion Matrix and accuracy
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
````

# HyperParameter Tuning

### Max Depth Tuning

````python
#Here we create a dataframe of max_depth in range 1-80 and checkm the accuracy score to each max_depth
````

````python
# GridSearch to find optimal max_depth
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

# number of Kfold
n_folds=5

# parameter to build the model

parameter={"max_depth":range(1,40)}

#instantiate the model
dtree=DecisionTreeClassifier(criterion="gini",random_state=100)


# fit tree the training data
tree=GridSearchCV(dtree,parameter,cv=n_folds,scoring="accuracy")

tree.fit(X_train,y_train)
````

````python
cv_results=tree.cv_results_
````

````python
cv_results=pd.DataFrame(cv_results)
cv_results.head()
````

````python
# Plotting accuracies with max_depth
plt.figure()
plt.plot(cv_results.param_max_depth,cv_results.mean_train_score,label="Training Accuracy")
plt.plot(cv_results.param_max_depth,cv_results.mean_test_score,label="Test Accuracy")
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
````

Thus, it is clear that the model is overfitting the training data if the max_depth is too high. Next, let's see how the model behaves with other hyperparameters.


###  Min_samples_leaf Tuning

````python
# Grid Search CV
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

# Specify number of folds for k-fold CV
n_folds=5

# Parameters
parameters={"min_samples_leaf":range(5,200,20)}

# instanteius the model
Dtree=DecisionTreeClassifier(criterion="gini",random_state=100)

# fit the model
tree=GridSearchCV(estimator=Dtree,param_grid=parameters,cv=n_folds,scoring="accuracy")
tree.fit(X_train,y_train)
````

````python
cv_results=tree.cv_results_
cv_results=pd.DataFrame(cv_results)
````

````python
cv_results.head()
````

````python
plt.figure()
plt.plot(cv_results.param_min_samples_leaf,cv_results.mean_test_score,label="Test_Accuracy")
plt.plot(cv_results.param_min_samples_leaf,cv_results.mean_train_score,label="Test Accuracy")
plt.xlabel("min_sample_leaf")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
````

At low values of min_samples_leaf, the tree gets a bit overfitted. At values > 100, however, the model becomes more stable and the training and test accuracy start to converge.

### Min_samples_split Tuning

 the minimum no. of samples required to split an internal node. Its default value is 2, which means that even if a node is having 2 samples it can be furthur divided into leaf nodes.

````python
# Grid SEarch 
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

# Kfolds
n_folds=5

# parameters to buikd
parameter={"min_samples_split":range(5,200,20)}

dtree=DecisionTreeClassifier(criterion="gini",random_state=100)

tree=GridSearchCV(estimator=dtree,param_grid=parameter,cv=n_folds,scoring="accuracy")
tree.fit(X_train,y_train)
````

````python
cv_results=tree.cv_results_
````

````python
cv_results=pd.DataFrame(cv_results)
cv_results.head()
````

````python
plt.figure()
plt.plot(cv_results.param_min_samples_split,cv_results.mean_test_score,label="Test_Accuracy")
plt.plot(cv_results.param_min_samples_split,cv_results.mean_train_score,label="Train_Accuracy")
plt.xlabel("min_sample_tree")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
````

This shows that as increase the min_samples_split, the tree overfits lesser since the model is less complex.
